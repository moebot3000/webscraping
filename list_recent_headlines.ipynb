# Web scraping headlines

The website has 75 pages of headlines. We want to collect them all by opening up the pages one at a time and scraping the headlines off the page. We'll start by making a list of the page numbers (1-75) we want the program to open, and we'll make sure those numbers are **string** types.

```python
page_numbers = str(np.arange(1, 75+1, 1).tolist())
```

For each page number in our list, we want the program to open up that webpage and scrape the headlines. Let's set up a **for loop** to do this for us.

```python
headlines = []

for page in page_numbers:
  site = requests.get('https://newslookup.com/us/newspapers/page'+page)
  site_html = site.text
  regex = 'target="_blank" >(.*?)</a>...<a class="'
  page_headlines = re.findall(regex, site_html)
  headlines.extend(page_headlines)
```

Let's look at our data and see if it is the kind of thing we were hoping for. You might notice some weird stuff (&#39, and that kind of thing) - these are the HTML codes for various punctuation marks.
